```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
fine_dust = pd.read_csv('data/fine_dust.csv', encoding = 'utf-8')
fine_dust.head()
```

```python
fine_dust.isnull().sum()
```

```python
fine_dust.info()
```

```python
del fine_dust['권역명']
```

```python
corr = fine_dust.corr()
corr
```

```python
corr['초미세먼지(㎍/㎥)'].sort_values(ascending=False)
```

```python
del fine_dust['오존(ppm)']
del fine_dust['아황산가스농도(ppm)']
```

```python
fine_dust = fine_dust.sort_values(by='측정일자', ascending=True)
fine_dust
```

```python
fine_dust.rename(columns={fine_dust.columns[0] : 'date',
                          fine_dust.columns[1] : 'gu',
                          fine_dust.columns[2] : 'pm10',
                          fine_dust.columns[3] : 'pm25',
                          fine_dust.columns[4] : 'no2',
                          fine_dust.columns[5] : 'co'}, inplace = True)
fine_dust.head()
```

```python
fine_dust = pd.pivot_table(fine_dust, index = ['gu', 'date'], values = ['pm10', 'pm25', 'no2', 'co'])
fine_dust
```

plt.figure(figsize = (15,5))
