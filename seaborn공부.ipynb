```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = 'darkgrid')

%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill', y = 'tip', data=tips)
```

hue 범례를 만들 수 있음

```python
sns.relplot(x='total_bill', y = 'tip', data=tips, hue='smoker')
```

```python
sns.relplot(x='total_bill', y = 'tip', data=tips,hue='size')
```

```python
sns.relplot(x='total_bill', y = 'tip', data=tips,style = 'smoker', hue='size')
```

```python
sns.relplot(x='total_bill', y = 'tip', data=tips,hue='size', palette ='Blues' )
```

```python
sns.relplot(x='total_bill', y = 'tip', data=tips,size='size' )
```
